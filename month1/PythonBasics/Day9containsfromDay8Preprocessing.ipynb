#  Day 9 – First Machine Learning Model

```python
x=df.drop("Survived",axis=1)
y=df["Survived"]
x.shape,y.shape
```

```python
x=x.drop("Ticket",axis=1)
x.shape
```

```python
print(x.columns)
```

```python
x=pd.get_dummies(x,columns=["Embarked"],drop_first=True)
```

```python
x.head()
```

```python
x.info()
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Train First Model(Logistic Regression)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
#Evaluate Model
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

Day-9 Summary:
Built first ML model using Logistic Regression.
Separated features and target, encoded categorical data,
split dataset into train and test sets, trained the model,
and evaluated performance using accuracy score.
