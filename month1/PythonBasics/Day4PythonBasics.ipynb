```python
#factorial(n) using loop
def factorial(n):
  fact=1
  for i in range(1,n+1):
    fact=fact*i
  return fact
n=int(input("Enter a number: "))
factorial(n)
```

```python
#largest(a, b, c)
def largest(a,b,c):
  if a>b and a>c:
    print(a,"is larger")
  elif b>c and b>a :
    print(b,"is larger")
  else:
    print(c,"is larger")
a=int(input("Enter a number: "))
b=int(input("Enter a number: "))
c=int(input("Enter a number: "))
largest(a,b,c)
```

```python
#count_vowels(string)
def count_vowels(s):
  n=len(s)
  c=0
  for i in range(n):
    for j in "aeiouAEIOU":
      if s[i]==j:
        c=c+1
  return c
s=input("Enter a string: ")
count_vowels(s)
```

```python
#Functions with Return Values
def list_stats(lst):
  n=len(lst)
  sum=0
  max=lst[0]
  for i in range(n):
    sum=sum+lst[i]
    if max<lst[i]:
      max=lst[i]
  avg=sum/n
  return sum,max,avg
lst=list(map(int,input().split()))
sum,max,average=list_stats(lst)
print("Sum=",sum,"Max=",max,"Average=",average)
```

```python
#is_prime(n)
def is_prime(n):
  c=0
  if n==1:
    return False
  for i in range(2,n):
    if n%i==0:
      c=c+1
  if c>0:
    return False
  else:
    return True
n=int(input("Enter a number: "))
is_prime(n)
```
