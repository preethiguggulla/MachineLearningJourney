```python
#Train-Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd

x_train['Sex'] = x_train['Sex'].map({"male": 0, "female": 1})
x_test['Sex'] = x_test['Sex'].map({"male": 0, "female": 1})

x_train = x_train.drop(columns=["PassengerId", "Name", "Ticket"], errors='ignore')
x_test = x_test.drop(columns=["PassengerId", "Name", "Ticket"], errors='ignore')


x_train = pd.get_dummies(x_train, columns=["Embarked"], drop_first=True)
x_test = pd.get_dummies(x_test, columns=["Embarked"], drop_first=True)

train_cols = x_train.columns
test_cols = x_test.columns

missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    x_test[c] = 0

missing_in_train = set(test_cols) - set(train_cols)
for c in missing_in_train:
    x_train[c] = 0

x_test = x_test[train_cols]

x_train = x_train.drop(columns=['Cabin', 'Emabrked'], errors='ignore')
x_test = x_test.drop(columns=['Cabin', 'Emabrked'], errors='ignore')

model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
#Evaluate Model
from sklearn.metrics import accuracy_score

x_test = x_test.fillna(0)

y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

Why accuracy alone is not enough?
Precision tells us: out of all passengers predicted as “survived”, how many actually survived.
Recall tells us: out of all passengers who actually survived, how many we correctly identified.
