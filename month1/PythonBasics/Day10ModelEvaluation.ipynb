```python
import pandas as pd
```

```python
df=pd.read_csv("/content/train.csv")
df
```

```python
df["Age"]=df["Age"].fillna(df["Age"].median())
df
```

```python
df["Emabrked"]=df["Embarked"].fillna(df["Embarked"].mode())
df
```

```python
x=df.drop("Survived",axis=1)
y=df["Survived"]
print(x)
print(y)
```

```python
x=x.drop(["PassengerId","Name","Ticket"],axis=1)
x.shape
```

```python
x.info()
```

```python
df["Sex"].value_counts(dropna=False)
```

```python
df["Sex"]=df["Sex"].astype(str)
df["Sex"]=df["Sex"].str.strip().str.lower()
df["Sex"]=df["Sex"].map({"male":0, "female":1})
df
```

```python
df["Sex"].value_counts(dropna=False)
```

```python
df.head()
```

```python
x=pd.get_dummies(x,columns=["Embarked"],drop_first=True)
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#
from sklearn.linear_model import LogisticRegression


x_test['Sex'] = x_test['Sex'].map({"male": 0, "female": 1})

x_train = x_train.drop(columns=['Cabin', 'Emabrked'], errors='ignore')
x_test = x_test.drop(columns=['Cabin', 'Emabrked'], errors='ignore')

model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
#Evaluate Model
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

Why accuracy alone is not enough?
Precision tells us: out of all passengers predicted as “survived”, how many actually survived.
Recall tells us: out of all passengers who actually survived, how many we correctly identified.
