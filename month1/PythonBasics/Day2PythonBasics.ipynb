```python
#Create a list of 5 numbers
#Add and remove items
#Print the first 3 items
n=int(input("How many items: "))
lst=[]
for i in range(n):
  item=input("Enter the item: ")
  lst.append(item)
print("List: ",lst)
lst.remove(input("Enter the item to remove:"))
print("List after removing an element: ",lst)
lst.append(input("Enter the item to add:"))
print("List after adding an element: ",lst)
print("Printing first three elements: ",lst[:3])
```

```python
#Create a tuple
nums=tuple(map(int,input("Enter elements seperated by space: ").split()))
print(nums)
#Convert tuple → list
lst=list(nums)
print("Convert tuple → list",lst)
```

```python
#Create a set
nums=set(map(int,input("Enter elements seperated by space: ").split()))
print(nums)
#Remove duplicates from a list using set()
items = input("Enter values separated by space: ").split()
unique = set(items)
print("Unique items: ",unique)
```
