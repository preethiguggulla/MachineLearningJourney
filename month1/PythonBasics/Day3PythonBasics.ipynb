```python
# Conditions
#Check even/odd
x=int(input("Enter a number: "))
if x%2==0:
  print("Even")
else:
  print("Odd")
```

```python
#Largest of 3 numbers
x=int(input("Enter first number: "))
y=int(input("Enter second number: "))
z=int(input("Enter third number: "))
if x>y and x>z :
  print(x,"is larger")
elif y>x and y>x :
  print(y,"is larger")
else :
  print(z,"is larger")
```

```python
#Grade calculator (A/B/C/D)
x=int(input("Enter your marks for 100: "))
if x>=90 and x<=100 :
  print("A grade")
elif x>=80 and x<90 :
  print("B grade")
elif x>=70 and x<80 :
  print("C grade")
elif x>=60 and x<70 :
  print("D grade")
else :
  print("Not valid")
```

```python
#Positive/negative/zero
x=int(input("Enter a number: "))
if x>0:
  print("Positive")
elif x<0 :
  print("Negative")
else:
  print("Zero")

```

```python
#Sum of numbers 1–100
sum=0
for i in range(1,101):
  sum=sum+i
print(sum)
```

```python
#Multiplication table (input from user)
n=int(input("Enter a number: "))
for i in range(1,11):
  x=n*i
  print(n,"x",i,"=",x)
```

```python
#Reverse a number using loop
x=int(input("Enter a number: "))
rev=0
while x>0:
  rem=x%10
  rev=rev*10+rem
  x=x//10
print(rev)
```

```python
#Combined logic
#Palindrome check
x=int(input("Enter a number: "))
rev=0
temp=x
while x>0:
  rem=x%10
  rev=rev*10+rem
  x=x//10
if temp==rev:
  print("Palindrome")
else:
  print("Not a Palindrome")
```

```python
#Factorial (loop)
x=int(input("Enter a number: "))
temp=1
for i in range(1,x+1):
  temp=temp*i
print("Factorial for",x,"is",temp)
```

```python
#Fibonacci (first 10 numbers)
x=10
a=0
b=1
print("Fibonacci series for first 10 numbers")
print(a)
print(b)
for i in range(x-2):
  c=a+b
  a=b
  b=c
  print(c)
```
